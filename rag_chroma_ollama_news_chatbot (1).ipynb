# Combining all topic-specific news lists into a single 'master_Data' list.

```python
master_Data = []

topics = [india_news,politics_news,movie_news,stocks_news,tech_news]
for topic in topics:
    master_Data.extend(topic)
```

```python
len(master_Data)
```

# Extracting news descriptions, metadata, and generating IDs for each article in master_Data.

```python
news = []
metadata = []
ids = []
for i,data in enumerate(master_Data):
    temp_dict = {"title":data.get('title'),
                 "published date":data.get('published date'),
                 "source":data["publisher"].get('title')
                 }
    ids.append(str(i+1))
    metadata.append(temp_dict)
    news.append(data.get('description'))
```

# Initializing ChromaDB, setting up the database path, and creating/getting the 'news_search' collection.

```python
import os
import chromadb
from chromadb.config import Settings
# DIR = os.path.dirname(os.path.abspath(__file__))
DIR = os.getcwd()
DB_PATH = os.path.join(DIR, 'data')
chroma_client = chromadb.PersistentClient(path=DB_PATH, settings=Settings(allow_reset=True, anonymized_telemetry=False))
news_search = chroma_client.get_or_create_collection(name="news_search")
```

# Adding news articles, metadata, and IDs to the 'news_search' collection.

```python
news_search.add(documents=news, metadatas=metadata, ids=ids)
```

# Defines a function to query the vector database and return relevant documents.

```python
def querying_vecdb(ques_input,total_chunks):
    query_result = news_search.query(query_texts= ques_input , n_results=total_chunks)
    result_documents = query_result["documents"][0]
    return result_documents

```

# Queries the vector database with a specific question and retrieves up to 50 relevant documents.

```python
ques_input = "Who won the champion trophy semi-final match?"
output = querying_vecdb(ques_input,50)
```

# Starts an Ollama server in a separate thread, making it accessible on all interfaces.

```python
import os
import threading
import subprocess

def ollama():
    os.environ['OLLAMA_HOST'] = '0.0.0.0:11434'
    os.environ['OLLAMA_ORIGINS'] = '*'
    subprocess.Popen(["ollama", "serve"])

ollama_thread = threading.Thread(target=ollama)
ollama_thread.start()
```

# Defines a `SimpleQA` class using LightRAG and Ollama to generate responses based on context.

```python
from lightrag.core.generator import Generator
from lightrag.core.component import Component
from lightrag.components.model_client import OllamaClient

qa_template = f"""<SYS>
You are a helpful chatbot assistant. Based on the below context give me an response
</SYS>
User: {conv_output}
You:"""

print(qa_template)

class SimpleQA(Component):
    def __init__(self, model_client, model_kwargs):
        super().__init__()
        self.generator = Generator(
            model_client=model_client,
            model_kwargs=model_kwargs,
            template=qa_template,
        )

    def call(self, input: dict) -> str:
        return self.generator.call({"input_str": str(input)})

model = {
    "model_client": OllamaClient(),
    "model_kwargs": {"model": "llama3.1:8b"}
}
qa = SimpleQA(**model)
```

# Time the execution of the call method of the qa object with the input ques_input and print the output

```python
%%time
output = qa.call(ques_input)
print(output)
```
